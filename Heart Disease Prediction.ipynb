```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
df_scale=scaler.fit_transform(df)
df1=pd.DataFrame(df_scale)
df1.head()
```

```python
x=df1.iloc[:,:-1].values
y=df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=0)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
clf = LogisticRegression(solver='liblinear') 
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_leaf=2, random_state=1)
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
     }   
```

```python
rf = RandomForestClassifier()
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

# KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier 
clf = KNeighborsClassifier(n_neighbors=8) 
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

```python
from sklearn.model_selection import cross_val_score
acc_rate=[]
for i in range(1,30):
    clf = KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(clf, x, y, cv=10)
    acc_rate.append(score.mean())
    
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,30), acc_rate, color='blue', linestyle='dashed')
plt.title("cross_val_score vs k-value")
plt.xlabel("k value")
plt.ylabel("accuracy rate")
```

```python
plt.figure(figsize=(20,10))
plt.plot([k for k in range(1,30)], acc_rate, color='blue', linestyle='dashed')
for i in range(1,30):
    plt.text(i, acc_rate[i-1], (i, acc_rate[i-1]))
    plt.xticks([i for i in range(1,30)])
plt.title("cross_val_score vs k-value")
plt.xlabel("k value")
plt.ylabel("accuracy rate")
```
